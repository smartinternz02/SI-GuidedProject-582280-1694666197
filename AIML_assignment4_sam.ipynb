```python
#oulier removal by replacement with median for TotalWorkingYears
q1=df.TotalWorkingYears.quantile(0.25)
q3=df.TotalWorkingYears.quantile(0.75)
print(q1)
print(q3)
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit=q3+1.5*IQR
upper_limit
```

```python
df.median() #50% quantile
```

```python
df=df[df.TotalWorkingYears<upper_limit]
```

```python
sns.boxplot(df.TotalWorkingYears)
```

```python
sns.boxplot(df.WorkLifeBalance)  #no outliers
```

```python
sns.boxplot(df.YearsInCurrentRole) #IQR method
```

```python
lower_percentile = 5
upper_percentile = 80

# Calculate the lower and upper bounds based on percentiles
lower_bound = df['YearsInCurrentRole'].quantile(lower_percentile / 100)
upper_bound = df['YearsInCurrentRole'].quantile(upper_percentile / 100)
```

```python
df_no_outliers = df[(df['YearsInCurrentRole'] >= lower_bound) & (df['YearsInCurrentRole'] <= upper_bound)]

# Create a boxplot to visualize the 'Age' column without outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_no_outliers, y='YearsInCurrentRole')
```

7.ENCODING

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x.BusinessTravel=le.fit_transform(x.BusinessTravel)
x.DailyRate=le.fit_transform(x.DailyRate)
x.Department=le.fit_transform(x.Department)
x.EducationField=le.fit_transform(x.EducationField)
x.Gender=le.fit_transform(x.Gender)
x.MaritalStatus=le.fit_transform(x.MaritalStatus)
x.JobRole=le.fit_transform(x.JobRole)
x.Over18=le.fit_transform(x.Over18)
x.OverTime=le.fit_transform(x.OverTime)
x.head()
```

```python
x.info()
```

8.FEATURE SCALING

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_scaled=pd.DataFrame(ms.fit_transform(x),columns=x.columns)
```

```python
x_scaled
```

9.SPLITTING OF DATA INTO TESTING AND TRAINING

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

```python
#Roc-auc
probability=model.predict_proba(x_test)[:,1]
probability
```

```python
from sklearn.metrics import roc_curve, auc

# Convert 'Yes' to 1 and 'No' to 0 in y_test
y_test_binary = y_test.map({'Yes': 1, 'No': 0})

# Calculate the probability scores
probability = model.predict_proba(x_test)[:, 1]

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test_binary, probability)

# Calculate AUC (Area Under the Curve)
roc_auc = auc(fpr, tpr)

# Now, you can plot the ROC curve using Matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
